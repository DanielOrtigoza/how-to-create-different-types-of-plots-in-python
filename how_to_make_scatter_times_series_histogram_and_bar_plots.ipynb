# Second example: time series

```python
days = np.linspace(start = 1, stop = 365, num = 365)

tmax = np.random.randint(low = 10, high = 40, size = 365)
tmin = np.random.randint(low = -30, high = -10, size = 365)

month_values = [1,32,61,92,122,153,183,214,245,275,306,336]
month_names = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

temperature_values = [-30,-20,-10,0,10,20,30,40]
temperature_names = ['-30 C','-20 C','-10 C','0 C','10 C','20 C','30 C','40 C']
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')
fig = plt.figure(3, facecolor = 'white', edgecolor = 'white', constrained_layout = True, figsize = (cm2inch(20), cm2inch(15)))
ax = plt.axes()
ax.set(facecolor = 'white', frame_on = False)

plt.plot(days, tmin, color = colors['orange1'], linewidth = 1, label = 'Minimum temperatures')
plt.plot(days, tmax, color = colors['green'], linewidth = 1, label = 'Maximum temperatures')

plt.fill_between(days, tmin, tmax, color = colors['blue'], alpha = 0.05)

plt.ylabel('Temperature C', fontsize = 'large')
plt.xlabel('Month', fontsize = 'large')
plt.title('Maximum and minimum temperatures', fontsize = 'large')
plt.legend(framealpha = 0.8, fontsize = 'medium', facecolor = 'white')
plt.grid(visible = True, which = "major", color = "#666666", linestyle = "-", alpha = 0.6)
plt.minorticks_on()
plt.grid(visible = True, which = "minor", color = "#999999", linestyle = "-", alpha = 0.2)

ax.set_xticks(month_values)
ax.set_xticklabels(month_names)

ax.set_yticks(temperature_values)
ax.set_yticklabels(temperature_names)

ax.set_xlim(1,336)

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

# Third example: histogram

In order to calculate the bin widths we are going to use the Freedman-Diaconis rule whic set the bid width to $b = \frac{max(n)-min(n)}{h}$ where $h = 2\ \mathrm{IQR}\ n^{-1/3}$ and $n$ is the number of observations. 

```python
# generating some data
shape, scale = 2, 2
data_hist = np.random.gamma(shape, scale, size = 3000)

# Freedman-Diaconis rule
q75, q25 = np.percentile(data_hist, [75 ,25])
iqr = q75 - q25
h = 2 * iqr * (len(data_hist) ** (-1/3))
b = (max(data_hist) - min(data_hist)) / h
b = int(round(b, 0))
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')
fig = plt.figure(5, facecolor = 'white', edgecolor = 'white', constrained_layout = True)#, figsize = (cm2inch(26.0), cm2inch(26.0)))
ax = plt.axes()
ax.set(facecolor = 'white', frame_on = False)

plt.hist(data_hist, bins = b, color = colors['green'], label = 'data', log = False, histtype = "stepfilled", alpha = 0.75)

min_ylim, max_ylim = plt.ylim()
    
plt.vlines(np.median(data_hist), 0, max_ylim * 0.9, color = colors['orange1'], label = 'Median: {:.2f}'.format(np.median(data_hist)))
plt.vlines(np.mean(data_hist), 0, max_ylim * 0.9, color = colors['orange2'], label = 'Mean: {:.2f}'.format(np.mean(data_hist))) 

plt.legend(framealpha = 0.8, fontsize = "medium", facecolor = "white")
plt.grid(visible = True, which = "major", color = "#666666", linestyle = "-", alpha = 0.6)
plt.minorticks_on()
plt.grid(visible = True, which = "minor", color = "#999999", linestyle = "-", alpha = 0.2)

plt.ylabel('y label', fontsize = 'large')
plt.xlabel('x label', fontsize = 'large')
plt.title('Title', fontsize = 'large')

plt.xlim([-0.5, max(data_hist)])

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

# Fourth example: bar charts

### One bar plot

```python
types_of_dogs = [
    'German\nShepherd',
    'Bulldog',
    'Labrador\nRetriever',
    'Golden\nRetriever',
    'Siberian\nHusky',
    'French\nBulldog',
    'Border\nCollie']
bar = np.arange(len(types_of_dogs))
data_bar = np.arange(len(bar)) + 1
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')
fig = plt.figure(7, facecolor = 'white', edgecolor = 'white', constrained_layout = True)#, figsize = (cm2inch(26.0), cm2inch(26.0)))
ax = plt.axes()
ax.set(facecolor = "white", frame_on = False)

width = 0.25
plt.bar(bar, data_bar, width, color = colors['teal'], align = 'center', label = 'teal label')

plt.legend(framealpha = 0.8, fontsize = 'medium', facecolor = 'white')
plt.ylabel('Number', fontsize = 'large')
plt.xlabel('Breeds', fontsize = 'large')
plt.title('Title', fontsize = 'large')

plt.xticks(bar, types_of_dogs, rotation = 45)

# to draw the text on the graph
for index, data in enumerate(data_bar):
    plt.text(x = index - 0.125, y = data + 0.1, s = '{:.1f}'.format(data), fontdict = dict(fontsize = 10))

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

### Two bar plot

```python
x_label = ['text_1','text_2','text_3','text_4','text_5']
data_bar1 = np.random.random(size = len(x_label))
data_bar2 = np.random.random(size = len(x_label))
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')
fig = plt.figure(8, facecolor = 'white', edgecolor = 'white', constrained_layout = True)#, figsize = (cm2inch(26.0), cm2inch(26.0)))
ax = plt.axes()
ax.set(facecolor = "white", frame_on = False)
ax.yaxis.set_major_formatter(mtick.PercentFormatter()) # format labels as percentaje

width = 0.25
bar1 = np.arange(len(x_label))
bar2 = [i - width for i in bar1]

plt.bar(bar1, data_bar1 * 100, width, color = colors['green'], align = 'edge', label = 'green label')
plt.bar(bar2, data_bar2 * 100, width, color = colors['orange1'], align = 'edge', label = 'orange label')

plt.legend(framealpha = 0.8, fontsize = 'medium', facecolor = 'white')
plt.ylabel('y label', fontsize = 'large')
plt.xlabel('x label', fontsize = 'large')
plt.title('Title', fontsize = 'large')

plt.xticks(bar1, x_label, rotation = 45)

# to draw the text on the graph
data_bar1_text = np.round(data_bar1 * 100, 1)
data_bar2_text = np.round(data_bar2 * 100, 1)

for index, data in enumerate(data_bar1_text):
    plt.text(x = index, y = data + 1, s = '{:.1f}%'.format(data), fontdict = dict(fontsize = 10))

for index, data in enumerate(data_bar2_text):
    plt.text(x = index - 0.25, y = data + 1, s = '{:.1f}%'.format(data), fontdict = dict(fontsize = 10))

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd
import numpy as np
```

```python
def cm2inch(cm: float) -> float:
    return cm * (1 / 2.54)
```

Through the the file presented below I use the style fivethirtyeight (this is just a personal choose), but, you can reverse and use the default one

- plt.style.use('default')
- plt.style.use('fivethirtyeight')

# Color pallet

A simple color pallet

```python
colors ={
    'red': '#EB1E2C',
    'orange1': '#FD6F30',
    'orange2': '#F9A729',
    'yellow': '#F9D23C',
    'green': '#5FBB68',
    'blue': '#00A5E3',
    'teal': '#64CDCC',
    'aqua': '#91DCEA',
    'purple': '#A4A4D5',
    'ligth_purple': '#BBC9E5',
}
 
```

### One bar plot

```python
types_of_dogs = [
    'German\nShepherd',
    'Bulldog',
    'Labrador\nRetriever',
    'Golden\nRetriever',
    'Siberian\nHusky',
    'French\nBulldog',
    'Border\nCollie']
bar = np.arange(len(types_of_dogs))
data_bar = np.arange(len(bar)) + 1
```

```python
fig = plt.figure(7, facecolor = 'white', edgecolor = 'white', constrained_layout = True)#, figsize = (cm2inch(26.0), cm2inch(26.0)))
ax = plt.axes()
ax.set(facecolor = "white", frame_on = False)

width = 0.25
plt.bar(bar, data_bar, width, color = colors['teal'], align = 'center', label = 'teal label')

plt.legend(framealpha = 0.8, fontsize = 'medium', facecolor = 'white')
plt.ylabel('Number', fontsize = 'large')
plt.xlabel('Breeds', fontsize = 'large')
plt.title('Title', fontsize = 'large')

plt.xticks(bar, types_of_dogs, rotation = 45)

# to draw the text on the graph
for index, data in enumerate(data_bar):
    plt.text(x = index - 0.125, y = data + 0.1, s = '{:.1f}'.format(data), fontdict = dict(fontsize = 10))

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

### Two bar plot

```python
x_label = ['text_1','text_2','text_3','text_4','text_5']
data_bar1 = np.random.random(size = len(x_label))
data_bar2 = np.random.random(size = len(x_label))
```

```python
fig = plt.figure(8, facecolor = 'white', edgecolor = 'white', constrained_layout = True)#, figsize = (cm2inch(26.0), cm2inch(26.0)))
ax = plt.axes()
ax.set(facecolor = "white", frame_on = False)
ax.yaxis.set_major_formatter(mtick.PercentFormatter()) # format labels as percentaje

width = 0.25
bar1 = np.arange(len(x_label))
bar2 = [i - width for i in bar1]

plt.bar(bar1, data_bar1 * 100, width, color = colors['green'], align = 'edge', label = 'green label')
plt.bar(bar2, data_bar2 * 100, width, color = colors['orange1'], align = 'edge', label = 'orange label')

plt.legend(framealpha = 0.8, fontsize = 'medium', facecolor = 'white')
plt.ylabel('y label', fontsize = 'large')
plt.xlabel('x label', fontsize = 'large')
plt.title('Title', fontsize = 'large')

plt.xticks(bar1, x_label, rotation = 45)

# to draw the text on the graph
data_bar1_text = np.round(data_bar1 * 100, 1)
data_bar2_text = np.round(data_bar2 * 100, 1)

for index, data in enumerate(data_bar1_text):
    plt.text(x = index, y = data + 1, s = '{:.1f}%'.format(data), fontdict = dict(fontsize = 10))

for index, data in enumerate(data_bar2_text):
    plt.text(x = index - 0.25, y = data + 1, s = '{:.1f}%'.format(data), fontdict = dict(fontsize = 10))

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

### Three bar plot

```python
pasta_types = [
    'Pipe\nRigate', 
    'Gemelli',
    'Farfalle',
    'Tortiglioni',
    'Fusillini',
    'Spaghetti',
    'Garganelli',
    'Penne\nLisce',
    'Sedan\nRigati',
    'Gnocchi',
    'Cellentani',
    'Mafalde\nCorte',
    'Maccheroni',
    'Fusilli\nBucati',
    'Conchiglie\nRigate',
    'Fusilli\nBucati\nLunghi',
    'Girandole',
    'Casarecce']

data_bar1_e3 = np.random.randint(low = 1, high = 30, size = len(pasta_types))
data_bar2_e3 = np.random.randint(low = 1, high = 40, size = len(pasta_types))
data_bar3_e3 = np.random.randint(low = 1, high = 20, size = len(pasta_types))
```

```python
fig = plt.figure(9, facecolor = 'white', edgecolor = 'white', constrained_layout = True, figsize = (cm2inch(25), cm2inch(14)))
ax = plt.axes()
ax.set(facecolor = "white", frame_on = False)

width_e3 = 0.25
bar2 = np.arange(len(pasta_types))
bar1 = [i - width_e3 for i in bar2]
bar3 = [i + width_e3 for i in bar2]

plt.bar(bar1, data_bar1_e3, width_e3, color = colors['orange2'], align = 'center', label = 'orange label')
plt.bar(bar2, data_bar2_e3, width_e3, color = colors['yellow'], align = 'center', label = 'yellow label')
plt.bar(bar3, data_bar3_e3, width_e3, color = colors['green'], align = 'center', label = 'green label')

plt.legend(framealpha = 0.8, fontsize = 'medium', facecolor = 'white')
plt.ylabel('y label', fontsize = 'large')
plt.xlabel('Types of pasta', fontsize = 'large')
plt.title('Title', fontsize = 'large')

plt.xticks(bar2, pasta_types, rotation = 45, fontdict = dict(fontsize = 10))

# to draw the text on the graph
for index, data in enumerate(data_bar1_e3):
    plt.text(x = index - 0.35, y = data + 0.1, s = '{:.0f}'.format(data), fontdict = dict(fontsize = 10))

for index, data in enumerate(data_bar2_e3):
    plt.text(x = index - 0.05, y = data + 0.1, s = '{:.0f}'.format(data), fontdict = dict(fontsize = 10))

for index, data in enumerate(data_bar3_e3):
    plt.text(x = index + 0.15, y = data + 0.1, s = '{:.0f}'.format(data), fontdict = dict(fontsize = 10))

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

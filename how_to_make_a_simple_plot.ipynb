```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def cm2inch(cm: float) -> float:
    return cm * (1 / 2.54)
```

# Color pallet

A simple color pallet

```python
colors ={
    'red': '#EB1E2C',
    'orange1': '#FD6F30',
    'orange2': '#F9A729',
    'yellow': '#F9D23C',
    'green': '#5FBB68',
    'blue': '#00A5E3',
    'teal': '#64CDCC',
    'aqua': '#91DCEA',
    'purple': '#A4A4D5',
    'ligth_purple': '#BBC9E5',
}
```

# First example

```python
sx1 = np.arange(0, 10, 0.01)
sy1 = np.sin(sx1)
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

plt.figure(1, figsize = (cm2inch(20), cm2inch(13)), constrained_layout = True) # [10,5] significa el tamaño de la figura [x, y]

plt.plot(sx1, sy1, '-', color = colors['orange1'], linewidth = 2.5, label = '$f(x) = \sin(x)$')

plt.ylabel('$y$', fontsize = 'large')
plt.xlabel('$x$', fontsize = 'large')
plt.legend(framealpha = 0.8, fontsize = 'large')

plt.grid(visible = True, which = 'major', color = '#666666', linestyle = '-', alpha = 0.6)
plt.minorticks_on()
plt.grid(visible = True, which = 'minor', color = '#999999', linestyle = '-', alpha = 0.2)
plt.tight_layout()

# plt.savefig('ajuste_lineal_punto_1_taller_2.jpg', dpi = 1200, quality = 95)

plt.show()
```

# Second example

```python
sx2 = np.arange(0, 10, 0.01)

sy2_1 = 1 + (sx2 ** 2)
sy2_2 = 1 + (sx2 ** 2) - (sx2 ** 3)
sy2_3 = 1 + (sx2 ** 2) - (sx2 ** 3) + (sx2 ** 4)
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

plt.figure(1, figsize = (cm2inch(20), cm2inch(13)), constrained_layout = True) # [10,5] significa el tamaño de la figura [x, y]

plt.plot(sx2, sy2_1, '-', color = colors['red'], linewidth = 2.5, label = '$f(x) = 1 + x^2$')
plt.plot(sx2, sy2_2, '--', color = colors['orange1'], linewidth = 2.5, label = '$f(x) = 1 + x^2 - x^3$')
plt.plot(sx2, sy2_3, '-', color = colors['orange2'], linewidth = 2.5, label = '$f(x) = 1 + x^2 - x^3 + x^4$')

plt.ylabel('$y$', fontsize = 'large')
plt.xlabel('$x$', fontsize = 'large')
plt.legend(framealpha = 0.8, fontsize = 'large')

plt.grid(visible = True, which = 'major', color = '#666666', linestyle = '-', alpha = 0.6)
plt.minorticks_on()
plt.grid(visible = True, which = 'minor', color = '#999999', linestyle = '-', alpha = 0.2)
plt.tight_layout()

# plt.savefig('ajuste_lineal_punto_1_taller_2.jpg', dpi = 1200, quality = 95)

plt.show()
```

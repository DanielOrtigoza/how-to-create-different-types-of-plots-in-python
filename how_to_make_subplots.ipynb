# First example

### Two subplots
Here the trick is set the labels correctly, and what i mean with that is setting the labels in a way that is centered and cover the whole subplot

```python
x = np.arange(10)
y = 2*x
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

fig, axes = plt.subplots(1, 2, facecolor = 'white', edgecolor = 'white', constrained_layout = True, sharey = True)#, figsize = (cm2inch(40), cm2inch(40)))

for i in axes:
    i.set(facecolor = 'white', frame_on = False)
    i.grid(visible = True, which = 'major', color = '#666666', linestyle = '-', alpha = 0.6)
    i.minorticks_on()
    i.grid(visible = True, which = 'minor', color = '#999999', linestyle = '-', alpha = 0.2)

ax1, ax2 = axes

ax1.plot(x, y, '.')
ax2.plot(x, x**2, '.')

# to set the labels
axes[0].set_ylabel('y label', fontsize = 'large')
axes[0].set_xlabel('x label', fontsize = 'large', color = 'white')
axes[0].set_title('title', fontsize = 'large', color = 'white')

fig.text(0.55, 0.03, 'x label', fontsize = 'large' , ha = 'center', va = 'center')
fig.text(0.55, 0.96, 'title', fontsize = 'large' , ha = 'center', va = 'center')

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

As you saw in this example, we have a (1,2) subplot, so, the y label is share by the two subplots without a problem. The problem began with the x label and the title, for those, the trick is setting individual x label and title for one of the subplots (in this case the first one (1,1)), with a white color, this, to make room for the real ones, which we drae with the .text

#### Same as above without the grid lines and with the data labels

# Second example

Here I'll show you hoy to set multiple histograms in a subplot, just to demonstrate a little trick which will help you plotting multiple plots with a single for loop.

In order to calculate the bin widths we are going to use the Freedman-Diaconis rule whic set the bid width to $b = \frac{max(n)-min(n)}{h}$ where $h = 2\ \mathrm{IQR}\ n^{-1/3}$ and $n$ is the number of observations. 

#### With Freedman-Diaconis rule

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

fig, axes = plt.subplots(2, 3, facecolor = 'white', edgecolor = 'white', constrained_layout = True, sharex = True, figsize = (cm2inch(30), cm2inch(15)))

((ax1,ax2,ax3), (ax4,ax5,ax6)) = axes

for i in axes:
    for j in range(len(i)):
        i[j].set(facecolor = 'white', frame_on = False)
        i[j].grid(visible = True)

axn = [ax1,ax2,ax3,ax4,ax5,ax6]

mu, sigma = 2, 0.1

for i in range(len(axn)):
    n = 10 ** (i + 1)
    data_hist = np.random.normal(mu, sigma, size = n)

    # Freedman-Diaconis rule
    q75, q25 = np.percentile(data_hist, [75 ,25])
    iqr = q75 - q25
    h = 2 * iqr * (len(data_hist) ** (-1/3))
    b = (max(data_hist) - min(data_hist)) / h
    b = int(round(b, 0))

    axn[i].hist(data_hist, bins = b, color = colors['green'], label = 'data', log = False, histtype = 'stepfilled', alpha = 0.75)
    axn[i].set_title('n = {}'.format(n))

ax4.set_ylabel('Frequency', fontsize = 'x-large', color = 'white')
fig.text(0.0125, 0.50, 'Frequency', fontsize = 'large' , ha = 'center', va = 'center', rotation = 90)

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

#### Without Freedman-Diaconis rule

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

fig, axes = plt.subplots(2, 3, facecolor = 'white', edgecolor = 'white', constrained_layout = True, sharex = True, figsize = (cm2inch(30), cm2inch(15)))

((ax1,ax2,ax3), (ax4,ax5,ax6)) = axes

for i in axes:
    for j in range(len(i)):
        i[j].set(facecolor = 'white', frame_on = False)
        i[j].grid(visible = True)

axn = [ax1,ax2,ax3,ax4,ax5,ax6]

mu, sigma = 2, 0.1

for i in range(len(axn)):
    n = 10 ** (i + 1)
    data_hist = np.random.normal(mu, sigma, size = n)

    axn[i].hist(data_hist, color = colors['green'], label = 'data', log = False, histtype = 'stepfilled', alpha = 0.75)
    axn[i].set_title('n = {}'.format(n))

ax4.set_ylabel('Frequency', fontsize = 'x-large', color = 'white')
fig.text(0.0125, 0.50, 'Frequency', fontsize = 'large' , ha = 'center', va = 'center', rotation = 90)

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

#### Let's look at different types of distributions into a single plot

```python
global_size = 10000

mu_n, sigma_n = 2, 0.1
normal_distribution = np.random.normal(mu_n, sigma_n, size = global_size)

n, p_b = 10000, 0.01
binomial_distribution = np.random.binomial(n, p_b, size = global_size)

shape, scale = 5, 2
gamma_distribution = np.random.gamma(shape, scale, size = global_size)

p_g = 0.01
geometric_distribution = np.random.geometric(p_g, size = global_size)

mu_l, sigma_l = 2, 0.1
lognormal_distribution = np.random.lognormal(mu_l, sigma_l, size = global_size)

lam = 40
poisson_distribution = np.random.poisson(lam, size = global_size)
```

```python
# Freedman-Diaconis rule normal distribution
q75_nd, q25_nd = np.percentile(normal_distribution, [75 ,25])
iqr_nd = q75_nd - q25_nd
h_nd = 2 * iqr_nd * (global_size ** (-1/3))
b_nd = (max(normal_distribution) - min(normal_distribution)) / h_nd
b_nd = int(round(b_nd, 0))

# Freedman-Diaconis rule binomial distribution
q75_bd, q25_bd = np.percentile(binomial_distribution, [75 ,25])
iqr_bd = q75_bd - q25_bd
h_bd = 2 * iqr_bd * (global_size ** (-1/3))
b_bd = (max(binomial_distribution) - min(binomial_distribution)) / h_bd
b_bd = int(round(b_bd, 0))

# Freedman-Diaconis rule gamma distribution
q75_gd, q25_gd = np.percentile(gamma_distribution, [75 ,25])
iqr_gd = q75_gd - q25_gd
h_gd = 2 * iqr_gd * (global_size ** (-1/3))
b_gd = (max(gamma_distribution) - min(gamma_distribution)) / h_gd
b_gd = int(round(b_gd, 0))

# Freedman-Diaconis rule geometric distribution
q75_ged, q25_ged = np.percentile(geometric_distribution, [75 ,25])
iqr_ged = q75_ged - q25_ged
h_ged = 2 * iqr_ged * (global_size ** (-1/3))
b_ged = (max(geometric_distribution) - min(geometric_distribution)) / h_ged
b_ged = int(round(b_ged, 0))

# Freedman-Diaconis rule locnormal distribution
q75_lnd, q25_lnd = np.percentile(lognormal_distribution, [75 ,25])
iqr_lnd = q75_lnd - q25_lnd
h_lnd = 2 * iqr_lnd * (global_size ** (-1/3))
b_lnd = (max(lognormal_distribution) - min(lognormal_distribution)) / h_lnd
b_lnd = int(round(b_lnd, 0))

# Freedman-Diaconis rule poisson distribution
q75_pd, q25_pd = np.percentile(poisson_distribution, [75 ,25])
iqr_pd = q75_pd - q25_pd
h_pd = 2 * iqr_pd * (global_size ** (-1/3))
b_pd = (max(poisson_distribution) - min(poisson_distribution)) / h_pd
b_pd = int(round(b_pd, 0))
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

fig, axes = plt.subplots(2, 3, facecolor = 'white', edgecolor = 'white', constrained_layout = True, figsize = (cm2inch(30), cm2inch(15))) # sharex = True

((ax1,ax2,ax3), (ax4,ax5,ax6)) = axes

for i in axes:
    for j in range(len(i)):
        i[j].set(facecolor = 'white', frame_on = False)
        i[j].grid(visible = True)

ax1.hist(normal_distribution, bins = b_nd, color = colors['purple'], histtype = "stepfilled", alpha = 0.75)
ax2.hist(binomial_distribution, bins = b_bd, color = colors['green'], histtype = "stepfilled", alpha = 0.75)
ax3.hist(gamma_distribution, bins = b_gd, color = colors['blue'], histtype = "stepfilled", alpha = 0.75)
ax4.hist(geometric_distribution, bins = b_ged, color = colors['red'], histtype = "stepfilled", alpha = 0.75)
ax5.hist(lognormal_distribution, bins = b_lnd, color = colors['orange1'], histtype = "stepfilled", alpha = 0.75)
ax6.hist(poisson_distribution, bins = b_pd, color = colors['orange2'], histtype = "stepfilled", alpha = 0.75)

ax1.set_ylabel('Frequency', fontsize = 'x-large', color = 'white')
fig.text(0.0125, 0.50, 'Frequency', fontsize = 'large' , ha = 'center', va = 'center', rotation = 90)

ax1.set_title('Normal distribution', fontsize = 'large')
ax2.set_title('Binomial distribution', fontsize = 'large')
ax3.set_title('Gamma distribution', fontsize = 'large')
ax4.set_title('geometric distribution', fontsize = 'large')
ax5.set_title('Lognormal distribution', fontsize = 'large')
ax6.set_title('Poisson distribution', fontsize = 'large')

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

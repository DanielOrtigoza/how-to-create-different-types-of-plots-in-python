# First example: subplots

### Two subplots
Here the trick is set the labels correctly, and what i mean with that is setting the labels in a way that is centered and cover the whole subplot

```python
x = np.arange(10)
y = 2*x
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

fig, axes = plt.subplots(1, 2, facecolor = 'white', edgecolor = 'white', constrained_layout = True, sharey = True)#, figsize = (cm2inch(40), cm2inch(40)))

for i in axes:
    i.set(facecolor = 'white', frame_on = False)
    i.grid(visible = True, which = 'major', color = '#666666', linestyle = '-', alpha = 0.6)
    i.minorticks_on()
    i.grid(visible = True, which = 'minor', color = '#999999', linestyle = '-', alpha = 0.2)

ax1, ax2 = axes

ax1.plot(x, y, '.')
ax2.plot(x, x**2, '.')

# to set the labels
axes[0].set_ylabel('y label', fontsize = 'large')
axes[0].set_xlabel('x label', fontsize = 'large', color = 'white')
axes[0].set_title('title', fontsize = 'large', color = 'white')

fig.text(0.55, 0.03, 'x label', fontsize = 'large' , ha = 'center', va = 'center')
fig.text(0.55, 0.96, 'title', fontsize = 'large' , ha = 'center', va = 'center')

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

As you saw in this example, we have a (1,2) subplot, so, the y label is share by the two subplots without a problem. The problem began with the x label and the title, for those, the trick is setting individual x label and title for one of the subplots (in this case the first one (1,1)), with a white color, this, to make room for the real ones, which we drae with the .text

#### Same as above without the grid lines and with the data labels

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

fig, axes = plt.subplots(1, 2, facecolor = 'white', edgecolor = 'white', constrained_layout = True, sharey = True)#, figsize = (cm2inch(40), cm2inch(40)))

for i in axes:
    i.set(facecolor = 'white', frame_on = False)
    i.grid(visible = True)

ax1, ax2 = axes

ax1.plot(x, y, '.')
ax2.plot(x, x**2, '.')

# to draw the text on the graph
for index, data in enumerate(y):
    ax1.text(x = index, y = data + 0.5, s = '{:.0f}'.format(data), fontdict = dict(fontsize = 10))

for index, data in enumerate(x**2):
    plt.text(x = index, y = data + 0.5, s = '{:.0f}'.format(data), fontdict = dict(fontsize = 10))

# to set the labels
axes[0].set_ylabel('y label', fontsize = 'large')
axes[0].set_xlabel('x label', fontsize = 'large', color = 'white')
axes[0].set_title('title', fontsize = 'large', color = 'white')

fig.text(0.55, 0.03, 'x label', fontsize = 'large' , ha = 'center', va = 'center')
fig.text(0.55, 0.96, 'title', fontsize = 'large' , ha = 'center', va = 'center')

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```

### Nine subplots

Just to demonstrate how to set the x and y labels, and the title correctly

```python
x = np.arange(10)
y = x
y2 = x[::-1]
```

```python
%matplotlib widget

plt.style.use('default')

fig, axes = plt.subplots(3, 3, facecolor = 'white', edgecolor = 'white', constrained_layout = True, sharex = True,sharey = True)#, figsize = (cm2inch(40), cm2inch(40)))

for i in axes:
    for j in range(len(axes)):
        i[j].set(facecolor = 'white', frame_on = True)
        i[j].grid(visible = False)
        i[j].grid(visible = True, which = 'major', color = '#666666', linestyle = '-', alpha = 0.6)
        i[j].minorticks_on()
        i[j].grid(visible = True, which = 'minor', color = '#999999', linestyle = '-', alpha = 0.2)

((ax1,ax2,ax3), (ax4,ax5,ax6), (ax7,ax8,ax9)) = axes 

ax1.plot(x, y, color = colors['red'])
ax2.plot(x, y2, color = colors['orange1'])
ax3.plot(x, y, color = colors['orange2'])
ax4.plot(x, y2, color = colors['yellow'])
ax5.plot(x, y, color = colors['green'])
ax6.plot(x, y2, color = colors['blue'])
ax7.plot(x, y, color = colors['teal'])
ax8.plot(x, y2, color = colors['aqua'])
ax9.plot(x, y, color = colors['purple'])

# to set the labels
ax1.set_ylabel('y label', fontsize = 'large', color = 'white')
ax7.set_xlabel('x label', fontsize = 'large', color = 'white')
ax1.set_title('title', fontsize = 'large', color = 'white')

fig.text(0.02, 0.55, 'y label', fontsize = 'large' , ha = 'center', va = 'center', rotation = 90)
fig.text(0.55, 0.03, 'x label', fontsize = 'large' , ha = 'center', va = 'center')
fig.text(0.55, 0.975, 'title', fontsize = 'large' , ha = 'center', va = 'center')

# plt.savefig('graph_name.jpg', dpi = 600, facecolor = fig.get_facecolor(), edgecolor='none')

plt.show()
```
